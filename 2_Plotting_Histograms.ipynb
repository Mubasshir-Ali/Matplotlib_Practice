# Plotting Histograms

```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
# Machine Learning students random records
ml_students_age = np.random.randint(18, 45, (100))
# Python students random records
py_students_age = np.random.randint(15, 40, (100))
```

```python
print(ml_students_age)
display(ml_students_age)
```

```python
print(py_students_age)
display(py_students_age)
```

```python
print(ml_students_age)
print(py_students_age)
```

```python
# Histogram Function Mean show record in bar
plt.hist(ml_students_age)
plt.show()
```

```python
# Histogram Function Mean show record in bar
plt.hist(py_students_age)
plt.show()
```

```python
# Histogram Function Mean show record in bar
plt.hist(ml_students_age)

plt.title("ML Students Age Histograms")
plt.xlabel("Students Age Category")
plt.ylabel("No. Students Age")
plt.show()
```

```python
# Histogram Function Mean show record in bar
plt.hist(py_students_age)

plt.title("Python Students Age Histograms")
plt.xlabel("Students Age Category")
plt.ylabel("No. Students Age")
plt.show()
```

```python
# Histogram Function Mean show record in bar
plt.hist(ml_students_age)

plt.title("ML Students Age Histograms")
plt.xlabel("Students Age Category")
plt.ylabel("No. Students Age")
plt.show()

# Histogram Function Mean show record in bar
plt.hist(py_students_age)

plt.title("Python Students Age Histograms")
plt.xlabel("Students Age Category")
plt.ylabel("No. Students Age")
plt.show()
```

```python
# Use bins method
bins = [15,20,25,30,35,40,45]

# Histogram Function Mean show record in bar
plt.hist(ml_students_age, bins)

plt.title("ML Students Age Histograms")
plt.xlabel("Students Age Category")
plt.ylabel("No. Students Age")
plt.show()
```

```python
# Use bins method
bins = [15,20,25,30,35,40,45]

# Histogram Function Mean show record in bar
plt.hist(ml_students_age, bins, rwidth = 0.8)

plt.title("ML Students Age Histograms")
plt.xlabel("Students Age Category")
plt.ylabel("No. Students Age")
plt.show()
```

```python
# Use bins method
bins = [15,20,25,30,35,40,45]

# Histogram Function Mean show record in bar
plt.hist(ml_students_age, bins, rwidth = 0.8, histtype = "step")

plt.title("ML Students Age Histograms")
plt.xlabel("Students Age Category")
plt.ylabel("No. Students Age")
plt.show()
```

```python
# Use bins method
bins = [15,20,25,30,35,40,45]

# Histogram Function Mean show record in bar
plt.hist(ml_students_age, bins, rwidth = 0.8, histtype = "bar")

plt.title("ML Students Age Histograms")
plt.xlabel("Students Age Category")
plt.ylabel("No. Students Age")
plt.show()
```

```python
# Use bins method
bins = [15,20,25,30,35,40,45]

# Histogram Function Mean show record in bar
plt.hist(ml_students_age, bins, rwidth = 0.8, histtype = "bar", orientation = "horizontal")

plt.title("ML Students Age Histograms")
plt.xlabel("Students Age Category")
plt.ylabel("No. Students Age")
plt.show()
```

```python
# Use bins method
bins = [15,20,25,30,35,40,45]

# Histogram Function Mean show record in bar
plt.hist(ml_students_age, bins, rwidth = 0.8, histtype = "bar", orientation = "horizontal", color = "m")

plt.title("ML Students Age Histograms")
plt.xlabel("Students Age Category")
plt.ylabel("No. Students Age")
plt.show()
```

```python
# Use bins method
bins = [15,20,25,30,35,40,45]

# Histogram Function Mean show record in bar
plt.hist(ml_students_age, bins, rwidth = 0.8, histtype = "bar", orientation = "horizontal", color = "m", label = "ML Students")

plt.title("ML Students Age Histograms")
plt.xlabel("Students Age Category")
plt.ylabel("No. Students Age")
plt.legend()
plt.show()
```

```python
# Use bins method
bins = [15,20,25,30,35,40,45]

# Histogram Function Mean show record in bar
plt.hist(ml_students_age, bins, rwidth = 0.8, histtype = "bar", orientation = "horizontal", color = "m", label = "ML Students")

# Histogram Function Mean show record in bar
plt.hist(py_students_age, bins, rwidth = 0.8, histtype = "bar", orientation = "horizontal", color = "y", label = "Python Students")

plt.title("ML Students Age Histograms")
plt.xlabel("Students Age Category")
plt.ylabel("No. Students Age")
plt.legend()
plt.show()
```

```python
# Use bins method
bins = [15,20,25,30,35,40,45]

# Histogram Function Mean show record in bar
plt.hist(ml_students_age, bins, rwidth = 0.8, histtype = "bar", orientation = "vertical", color = "m", label = "ML Students")

# Histogram Function Mean show record in bar
plt.hist(py_students_age, bins, rwidth = 0.8, histtype = "bar", orientation = "vertical", color = "y", label = "Python Students")

plt.title("ML & Python Students Age Histograms")
plt.xlabel("Students Age Category")
plt.ylabel("No. Students Age")
plt.legend()
plt.show()
```

```python
# Use bins method
bins = [15,20,25,30,35,40,45]
# write Ml & Python Plot in single line
# Histogram Function Mean show record in bar
plt.hist([ml_students_age, py_students_age], bins, rwidth = 0.8, histtype = "bar", orientation = "vertical", color = ["m", "y"], label = ["ML Students", "Python Students"])

# Histogram Function Mean show record in bar
#plt.hist(py_students_age, bins, rwidth = 0.5, histtype = "bar", orientation = "vertical", color = "y", label = "Python Students")

plt.title("ML & Python Students Age Histograms")
plt.xlabel("Students Age Category")
plt.ylabel("No. Students Age")
plt.legend()
plt.show()
```
