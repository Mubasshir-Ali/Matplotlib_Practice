# Scatter Plot

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df_google_play_store_apps = pd.read_csv("google-play-store-apps/googleplaystore.csv")

df_google_play_store_apps.shape
```

```python
df_google_play_store_apps = pd.read_csv("google-play-store-apps/googleplaystore.csv", nrows = 1000)

df_google_play_store_apps.shape
```

```python
df_google_play_store_apps
```

```python
display(df_google_play_store_apps)
```

```python
x = df_google_play_store_apps["Rating"]
y = df_google_play_store_apps["Reviews"]
```

```python
# scatter plot
plt.scatter(x,y)
```

```python
# scatter plot
plt.scatter(x,y)
plt.title("Google Ply Store Apps Scatter Plot")
```

```python
# scatter plot
plt.scatter(x,y)
plt.title("Google Ply Store Apps Scatter Plot")
plt.xlabel("Rating")
```

```python
# scatter plot
plt.scatter(x,y)
plt.title("Google Ply Store Apps Scatter Plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
```

```python
# scatter plot
plt.scatter(x,y, c = 'r')
plt.title("Google Ply Store Apps Scatter Plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
```

```python
plt.figure(figsize = (16, 9))
# scatter plot
plt.scatter(x,y, c = 'r')
plt.title("Google Ply Store Apps Scatter Plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
```

```python
plt.figure(figsize = (16, 9))
# scatter plot
plt.scatter(x,y, c = 'r', marker = "*")
plt.title("Google Ply Store Apps Scatter Plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
```

```python
plt.figure(figsize = (16, 9))
# scatter plot
plt.scatter(x,y, c = 'r', marker = "*", s = 100)
plt.title("Google Ply Store Apps Scatter Plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
```

```python
plt.figure(figsize = (16, 9))
# scatter plot
plt.scatter(x,y, c = 'r', marker = "*", s = 100, alpha = 0.5)
plt.title("Google Ply Store Apps Scatter Plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
```

```python
plt.figure(figsize = (16, 9))
# scatter plot
plt.scatter(x,y, c = 'r', marker = "*", s = 100, alpha = 0.5, linewidths = 10)
plt.title("Google Ply Store Apps Scatter Plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
```

```python
plt.figure(figsize = (16, 9))
# scatter plot
plt.scatter(x,y, c = 'r', marker = "*", s = 100, alpha = 0.5, linewidths = 10, edgecolors = "g")
plt.title("Google Ply Store Apps Scatter Plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
```

```python
plt.figure(figsize = (16, 9))
# scatter plot
plt.scatter(x,y, c = 'r', marker = "*", s = 100, alpha = 0.5, linewidths = 10, edgecolors = "g", verts = "<")
plt.title("Google Ply Store Apps Scatter Plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
```

```python
plt.figure(figsize = (16, 9))
# scatter plot
plt.scatter(x,y, c = 'r', marker = "*", s = 100, alpha = 0.5, linewidths = 10, edgecolors = "g")
plt.scatter(x,df_google_play_store_apps["Installs"], c = 'y', marker = "p", s = 100, alpha = 0.5, linewidths = 10, edgecolors = "c")


plt.title("Google Ply Store Apps Scatter Plot")
plt.xlabel("Rating")
plt.ylabel("Reviews & Installs")
```
