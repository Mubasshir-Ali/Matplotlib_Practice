# Bar Chart

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
```

```python
classes = ["Python", "R", "AI", "ML", "DS"]

# Out of 100 stuents in each class
class1_students = [30, 10, 20, 25, 10]
class2_students = [40, 5, 20, 20, 10]
class3_students = [35, 5, 30, 15, 15]
```

```python
plt.bar(classes, class1_students)
```

```python
plt.bar(classes, class2_students)
```

```python
plt.bar(classes, class3_students)
```

```python
# show horizantal plot using barh
plt.barh(classes, class1_students)
```

```python
plt.bar(classes, class1_students, width = 0.2)
```

```python
plt.bar(classes, class1_students, width = 0.2, align = "edge")
```

```python
plt.bar(classes, class1_students, width = 0.2, align = "edge", color = "y")
```

```python
plt.bar(classes, class1_students, width = 0.2, align = "edge", color = "y", edgecolor = "m")
```

```python
plt.bar(classes, class1_students, width = 0.2, align = "edge", color = "y", edgecolor = "m", linewidth = 5)
```

```python
plt.bar(classes, class1_students, width = 0.2, align = "edge", color = "y", edgecolor = "m", linewidth = 5, alpha = 0.9)
```

```python
plt.bar(classes, class1_students, width = 0.2, align = "edge", color = "y", edgecolor = "m", linewidth = 5, alpha = 0.9, linestyle = "--")
```

```python
plt.bar(classes, class1_students, width = 0.2, align = "edge", color = "y", edgecolor = "m", linewidth = 5, alpha = 0.9, linestyle = "--", label = "Class 1 Stulents")
plt.legend()
```

```python
plt.figure(figsize = (16, 9))
plt.bar(classes, class1_students, width = 0.2, align = "edge", color = "b", label = "Class 1 Stulents")
plt.bar(classes, class2_students, width = 0.2, align = "edge", color = "y", label = "Class 2 Stulents")
plt.bar(classes, class3_students, width = 0.2, align = "edge", color = "m", label = "Class 3 Stulents")

plt.legend()

plt.title("Bar Chart Of Class 1 & 2 & 3", fontsize = 20)
plt.xlabel("Classes", fontsize = 14)
plt.ylabel("No. Of Students", fontsize = 14)
plt.show()
```

```python
plt.figure(figsize = (16, 9))

classes_index = np.arange(len(classes))

plt.bar(classes_index, class1_students, width = 0.2, align = "edge", color = "b", label = "Class 1 Stulents")
plt.bar(classes_index, class2_students, width = 0.2, align = "edge", color = "y", label = "Class 2 Stulents")
plt.bar(classes_index, class3_students, width = 0.2, align = "edge", color = "m", label = "Class 3 Stulents")

plt.legend()

plt.title("Bar Chart Of Class 1 & 2 & 3", fontsize = 20)
plt.xlabel("Classes", fontsize = 14)
plt.ylabel("No. Of Students", fontsize = 14)
plt.show()
```

```python
plt.figure(figsize = (16, 9))

classes_index = np.arange(len(classes))
width = 0.2

plt.bar(classes_index, class1_students, width, align = "edge", color = "b", label = "Class 1 Stulents")
plt.bar(classes_index + width, class2_students, width, align = "edge", color = "y", label = "Class 2 Stulents")
plt.bar(classes_index + width + width, class3_students, width, align = "edge", color = "m", label = "Class 3 Stulents")

plt.legend()

plt.title("Bar Chart Of Class 1 & 2 & 3", fontsize = 20)
plt.xlabel("Classes", fontsize = 14)
plt.ylabel("No. Of Students", fontsize = 14)
plt.show()
```

```python
plt.figure(figsize = (16, 9))

classes_index = np.arange(len(classes))
width = 0.2

plt.bar(classes_index, class1_students, width, align = "edge", color = "b", label = "Class 1 Stulents")
plt.bar(classes_index + width, class2_students, width, align = "edge", color = "y", label = "Class 2 Stulents")
plt.bar(classes_index + width + width, class3_students, width, align = "edge", color = "m", label = "Class 3 Stulents")

plt.legend()

# xticks means show label in x axis because bydefault label is 0,1,2,3 etc
plt.xticks(classes_index + width, classes)
plt.title("Bar Chart Of Class 1 & 2 & 3", fontsize = 20)
plt.xlabel("Classes", fontsize = 14)
plt.ylabel("No. Of Students", fontsize = 14)
plt.show()
```

```python
plt.figure(figsize = (16, 9))

classes_index = np.arange(len(classes))
width = 0.2

plt.bar(classes_index, class1_students, width, align = "edge", color = "b", label = "Class 1 Stulents")
plt.bar(classes_index + width, class2_students, width, align = "edge", color = "y", label = "Class 2 Stulents")
plt.bar(classes_index + width + width, class3_students, width, align = "edge", color = "m", label = "Class 3 Stulents")

plt.legend()

# xticks means show label in x axis because bydefault label is 0,1,2,3 etc
plt.xticks(classes_index + width, classes, rotation = 20)
plt.title("Bar Chart Of Class 1 & 2 & 3", fontsize = 20)
plt.xlabel("Classes", fontsize = 14)
plt.ylabel("No. Of Students", fontsize = 14)
plt.show()
```

```python
plt.figure(figsize = (16, 9))

classes_index = np.arange(len(classes))
width = 0.2

plt.barh(classes_index, class1_students, width, align = "edge", color = "b", label = "Class 1 Stulents")
plt.barh(classes_index + width, class2_students, width, align = "edge", color = "y", label = "Class 2 Stulents")
plt.barh(classes_index + width + width, class3_students, width, align = "edge", color = "m", label = "Class 3 Stulents")

plt.legend()

# yticks means show label in y axis because bydefault label is 0,1,2,3 etc
plt.yticks(classes_index + width, classes, rotation = 20)
plt.title("Bar Chart Of Class 1 & 2 & 3", fontsize = 20)
plt.xlabel("Classes", fontsize = 14)
plt.ylabel("No. Of Students", fontsize = 14)
plt.show()
```
